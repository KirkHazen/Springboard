**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
###trying a function
##dfg = clean_prop.groupby(clean_prop['borough'])
dfg = clean_prop.groupby(by = ['borough', 'Year']).mean()
dfg.head()
```

```python
##lost the row index so
dfg = dfg.reset_index()
dfg.head()
```

```python
###Now for years and ratios
def price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
###testing it out with one
price_ratio(dfg[dfg['borough']=='Harrow'])
```

```python
###empty dictionary to start
df_final = {}
```

```python
###now to run it through all of them
for b in dfg['borough'].unique():
    borough = dfg[dfg['borough'] == b]
    df_final[b] = price_ratio(borough)
print(df_final)
```

```python
###turn df_final into an actual dataframe
ratios = pd.DataFrame(df_final)
ratios.head()
```

```python
###flipping and setting index
ratios_T = ratios.T
ratios_T = ratios_T.reset_index()
ratios_T.head()
```

```python
ratios_T.rename(columns={'index':'Borough', 0:'2018/1998'}, inplace=True)
ratios_T.head()
```

```python
biggest_movers = ratios_T.sort_values(by='2018/1998',ascending=False).head(20)
print(biggest_movers)
```

```python
ax = biggest_movers[['Borough','2018/1998']].plot(kind='bar')
ax.set_ylabel('Ratio of Price 2018/1998')
ax.set_xticklabels(biggest_movers.Borough)
###I do have a question here about the syntax to get the Borough column out of the dataframe. Why the .Borough
### and not biggest_movers['Borough']
```

```python
###Those that changed the least
least_movers = ratios_T.sort_values(by='2018/1998',ascending=False).tail(10)
print(least_movers)
```

```python
ax = least_movers[['Borough','2018/1998']].plot(kind='bar')
ax.set_ylabel('Ratio of Price 2018/1998')
ax.set_xticklabels(least_movers.Borough)
```
